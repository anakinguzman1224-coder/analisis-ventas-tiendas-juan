## **5. Agregando ID a mis DF y Consolidando las cuatro tiendas**

```python
tienda_df['ID_Tienda'] = 'Tienda'
tienda_df2['ID_Tienda'] = 'Tienda 2'
tienda_df3['ID_Tienda'] = 'Tienda 3'
tienda_df4['ID_Tienda'] = 'Tienda 4'
```

### **Concatenar para Crear el df_maestro, se une las cuatro bases de datos que ya tienen su identificador.**

```python
listas_dfs = [tienda_df, tienda_df2, tienda_df3, tienda_df4]
df_maestro = pd.concat(listas_dfs, ignore_index=True)
print(df_maestro.head())

print("¡DataFrame Maestro creado exitosamente!")
print(df_maestro.head())
print(f"\nTotal de filas consolidadas: {len(df_maestro)}")
```

```python
ventas_por_producto = df_maestro.groupby(['ID_Tienda', 'Producto']).size().reset_index(name='Cantidad_Vendida')
ventas_por_producto.tail()
```

```python
ventas_por_producto_ordenado = ventas_por_producto.sort_values(by=['ID_Tienda','Cantidad_Vendida'],ascending=[True, False])
ventas_por_producto_ordenado.head()
```

```python
print("--- Producto Más Vendido por Cantidad en cada Tienda ---")
# Usamos drop_duplicates() para mantener solo el primer producto (el más vendido) de cada grupo 'ID_Tienda'
top_ventas = ventas_por_producto_ordenado.drop_duplicates(subset='ID_Tienda', keep='first')
print(top_ventas)
```

```python
print("Agrupar por Tienda Y Producto, y SUMAR el Precio Ingreso")
ingresos_por_producto = df_maestro.groupby(['ID_Tienda', 'Producto'])['Precio'].sum().reset_index()
ingresos_por_producto.columns = ['ID_Tienda', 'Producto', 'Ingreso_Generado']

# 2. Encontrar el producto MÁS RENTABLE (mayor ingreso) por TIENDA
ingresos_ordenados = ingresos_por_producto.sort_values(
    by=['ID_Tienda', 'Ingreso_Generado'],
    ascending=[True, False] # True para ID_Tienda, False para Ingreso (mayor a menor)
)

# 3. Mostrar el más rentable de cada tienda
top_rentables = ingresos_ordenados.drop_duplicates(subset='ID_Tienda', keep='first')
print("\n--- Producto Más Rentable por Ingreso en cada Tienda ---")
print(top_rentables)

# Para el MENOS Rentable: ordenamos por Ingreso_Generado ascendentemente
menos_rentables = ingresos_por_producto.sort_values(by='Ingreso_Generado', ascending=True).drop_duplicates(subset='ID_Tienda', keep='first')
print("\n--- Producto Menos Rentable por Ingreso en cada Tienda ---")
print(menos_rentables)
```

### **Calcular ingresos por productos (CONSOLIDADO)**

```python
# 1. CALCULAR INGRESOS POR PRODUCTO (CONSOLIDADO)
ingresos_por_producto = df_maestro.groupby('Producto')['Precio'].sum().reset_index()
ingresos_por_producto.columns = ['Producto', 'Ingreso_Generado']

# 2. SELECCIONAR EL TOP 5 MUNDIAL
top_5_productos = ingresos_por_producto.sort_values(
    by='Ingreso_Generado', ascending=False
).head(5)
print("\n--- Top 5 Productos por Ingreso Generado ---")
print(top_5_productos)


```

### **6.3 Graficando Rentabilidad vs Cantidad Vendida (Gráfico de Barras Agrupadas con Doble Eje Y)**

```python
# Calcular las métricas clave en un solo comando
resumen_rentabilidad_df = df_maestro.groupby('ID_Tienda').agg(
    Cantidad_Total=('Precio', 'size'),  # Cuántas transacciones hubo
    Ingreso_Total=('Precio', 'sum'),      # Ingreso total
    Ingreso_Promedio=('Precio', 'mean')   # Ingreso promedio por transacción
).reset_index()

# Ordenar por el Ingreso Total (el factor de venta)
resumen_rentabilidad_df = resumen_rentabilidad_df.sort_values(by='Ingreso_Total', ascending=False)
```

```python
resumen_rentabilidad_df.head()
```

```python
# Crear la figura y los ejes
fig, ax1 = plot.subplots(figsize=(10, 6))

# --- Eje 1 (Izquierda): Ingreso Total (Barras Altas) ---
color_ingreso = 'tab:blue'
ax1.set_xlabel('ID de la Tienda')
ax1.set_ylabel('Ingreso Total (Millones $)', color=color_ingreso)
barras_ingreso = ax1.bar(
    resumen_rentabilidad_df['ID_Tienda'],
    resumen_rentabilidad_df['Ingreso_Total'],
    color=color_ingreso,
    alpha=0.6,
    label='Ingreso Total'
)
ax1.tick_params(axis='y', labelcolor=color_ingreso)

# --- Eje 2 (Derecha): Cantidad de Transacciones (Línea) ---
# Usamos un segundo eje para la métrica de cantidad (para que no distorsione la escala)
ax2 = ax1.twinx()
color_cantidad = 'tab:red'
ax2.set_ylabel('Cantidad Total de Transacciones', color=color_cantidad)
linea_cantidad, = ax2.plot(
    resumen_rentabilidad_df['ID_Tienda'],
    resumen_rentabilidad_df['Cantidad_Total'],
    color=color_cantidad,
    marker='o', # Puntos en la línea
    linestyle='-',
    linewidth=2,
    label='Cantidad Total'
)
ax2.tick_params(axis='y', labelcolor=color_cantidad)

# Añadir títulos y leyenda
plot.title('2. Ingreso Total vs. Cantidad de Transacciones por Tienda', fontsize=16)

# Asegurar que las leyendas de ambos ejes se muestren
lineas = [barras_ingreso[0], linea_cantidad]
etiquetas = [l.get_label() for l in lineas]
ax1.legend(lineas, etiquetas, loc='upper left')

fig.tight_layout()
plot.show()
```
